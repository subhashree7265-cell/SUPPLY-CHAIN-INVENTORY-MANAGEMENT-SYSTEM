```python
inventory = {
    "Laptop":10,
    "Mouse":25,
    "Keyboard":15
}
def view_inventory():
    print("\n---Inventory Dashboard---")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")

def restock_product():
    product = input("Enter product name: ")
    quantity = int(input("Enter quantity to add: "))

    if product in inventory:
        inventory[product] += quantity
    else:
        inventory[product] =  quantity

    print("stock updated sucessfully!")


def process_sale():
    product = input("Enter product name: ")
    quantity = int(input("Enter quantity sold: "))


    if product in inventory:
        if inventory[product] >= quantity:
          inventory[product] -= quantity
          print("Sale processed sucessfully!")
        else:
            print("Insufficient stock!")
    else:
        print("product not found!")


while True:
    print("\n===== Supply Chain Management =====")
    print("1. View Inventory Dashboard")
    print("2. Restock/Add Product")
    print("3. Process Order Sale")
    print("4. Exit")


    choice = input("Enter choice: ")

    if choice == "1":
        view_inventory()
    elif choice == "2":
        restock_product()
    elif choice == "3":
        process_sale()
    elif choice == "4":
        print("Exiting System...")
        break
    else:
        print("Invalid Choice!")
    
    
```
